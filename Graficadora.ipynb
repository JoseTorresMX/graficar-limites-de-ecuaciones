```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, FuncFormatter
```

```python
def y_func(x,t):
    return np.sin(t*x)/x
```

```python
x_valores=np.linspace(-2*np.pi, 2*np.pi, 400)
```

```python
t_valores=[1,2,3]
```

```python
plt.figure(figsize=(10, 6))
for t in t_valores:
    y_valores = y_func(x_valores, t)
    plt.plot(x_valores, y_valores, label=f't = {t}')
```

```python
x_limite = 0
y_limite_valores = [
    np.sin(t*x_limite)/x_limite if x_limite != 0 else 1 for t in t_valores]
```

```python
plt.scatter([x_limite]*len(t_valores), y_limite_valores,
            color='red', label='Limite en x=0')
```

```python
plt.title("Grafias")
plt.xlabel('x')
plt.ylabel('y')
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           ['-2', '-', '0', '', '2'])
plt.ylim(-1,3)
plt.legend()
plt.grid(True)
```

```python
plt.show()
```
